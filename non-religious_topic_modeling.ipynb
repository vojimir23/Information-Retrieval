### Tweets cleaning

```python
# NLTK Stop words
stop_words = stopwords.words('english')
stop_words.extend(["be","u","get", "say", "make", "think","tweet","retweet","twitter","many","today","time","thing","good","humanists"])
```

```python
def clean_tweet(tweet):
    punctuation = string.punctuation + '”'
    tweet = tweet.lower()
    tweet = contractions.fix(tweet)
    tweet = re.sub(r'^b(?:"|\')?(.*)$', r'\1', tweet)
    tweet = re.sub(r"\\n", '', tweet)
    tweet = re.sub(r'http\S+', '', tweet)
    tweet = re.sub(r'#\s*\w+', '', tweet)
    tweet = re.sub(r'@\s*\w+', '', tweet)
    tweet = re.sub(r'\\[a-z0-9]{3}', '', tweet)
    tweet = re.sub(r'\s+', ' ', tweet)  # remove double white spaces   
    tweet = tweet.translate(tweet.maketrans('', '', punctuation))
    tweet = ' '.join([word for word in tweet.split() if word not in stop_words])
    tweet = ' '.join([word for word in tweet.split() if len(word) > 3 or word in ["no", "not", "god", "end", "yes"]])
    return tweet
```

```python
def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

#def make_trigrams(texts):
    #return [trigram_mod[bigram_mod[doc]] for doc in texts]
nlp = spacy.load("en_core_web_sm",disable=['parser', 'ner'])
def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    texts_out = [] #lemmatized_tweets 
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out

# Form Bigrams
data_words_bigrams = make_bigrams(words)
#data_words_trigrams=make_trigrams(words)
#['NOUN', 'ADJ', 'VERB', 'ADV'])
data_lemmatized = lemmatization(data_words_bigrams, allowed_postags=['NOUN','ADJ'])
data_lemmatized = [lst for lst in data_lemmatized if len(lst) > 2] #del short lists
print(data_lemmatized[100])
```

```python
from gensim.models import CoherenceModel
from gensim.corpora import Dictionary
from gensim.models.ldamodel import LdaModel

# Define the range of hyperparameters to search over
params = {
    'num_topics': range(20, 65, 5),
    'alpha': np.arange(0.001, 1, 0.001).tolist(),
    'eta': np.arange(0.001, 0.5, 0.001).tolist(),
}

# Define the number of iterations to run
num_iters = 300

# Initialize lists to store results
coherence_scores = []
perplexity_scores = []
results = []

# Train LDA model 
for i in range(num_iters):
    # Sample hyperparameters randomly
    sampled_params = {k: np.random.choice(v) for k, v in params.items()}

    # Train LDA model with the sampled hyperparameters
    lda_model = LdaModel(
        corpus=corpus,
        id2word=id2word,
        num_topics=sampled_params['num_topics'],
        update_every=1,
        passes=2,
        alpha=sampled_params['alpha'],
        eta=sampled_params['eta'],
        eval_every=None,
        random_state=100,
        iterations=1000,
        per_word_topics=True)

    # Compute coherence score and perplexity of the trained model
    coherence_model_lda_cvr = CoherenceModel(model=lda_model, corpus=corpus, dictionary=id2word, texts=data_lemmatized, coherence='c_v')
    coherence_cvr = coherence_model_lda_cvr.get_coherence() # from 0 to 1
    perplexity = lda_model.log_perplexity(corpus)

    # Store results
    results.append({'num_topics': sampled_params['num_topics'], 'alpha': sampled_params['alpha'], 'eta': sampled_params['eta'], 'coherence_cvr': coherence_cvr, 'perplexity': perplexity})

    # Update progress
    print(f"Iteration {i+1}: params={sampled_params}, coherence_cvr={coherence_cvr:.4f}, perplexity={perplexity:.4f}")

# Convert results to a pandas dataframe
results_df = pd.DataFrame(results)
results_df.to_excel('results_final.xlsx', index=False)
```

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start, step):

    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                id2word=dictionary,
                                                num_topics=num_topics, 
                                                random_state=100,
                                                update_every=1,
                                                #chunksize=1000,
                                                passes=2,
                                                alpha=0.429,     #simetric=0.01 and asimetric? 0.001
                                                eta=0.464,
                                                eval_every=None,
                                                iterations=10000,
                                                per_word_topics=True)
        model_list.append(model)
        coherence_model_lda_cvr = CoherenceModel(model=model, corpus=corpus, dictionary=dictionary, texts=texts, coherence='c_v')
        coherence_values.append(coherence_model_lda_cvr.get_coherence())

    return model_list, coherence_values

model_list, coherence_values = compute_coherence_values(dictionary=id2word, corpus=corpus, texts=data_lemmatized, start=55, limit=66, step=1)

# Show graph
limit = 66
start = 55
step = 1
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
#plt.savefig("coherence_plot_final.png", dpi=500)
plt.show()
```

```python
tweets_df = df[["tweet"]]
# extract the probabilities of each topic for each tweet
tweets_topics = []
indices_to_remove = []
for index, row in tweets_df.iterrows():
    # extract the text of the tweet
    tweet = row['tweet']
    # convert the tweet to a bag-of-words vector
    bow_vector = lda_model.id2word.doc2bow(tweet.split())
    # get the probability of each topic for the tweet
    topic_probs = lda_model.get_document_topics(bow_vector)
    # create a list to store the probability of each topic for the tweet
    tweet_topics = [0] * lda_model.num_topics
    # iterate over each topic and its probability
    for topic, prob in topic_probs:
        # store the probability of the topic in the tweet_topics list
        tweet_topics[topic] = prob
    # check if all the topic probabilities are equal
    if len(set(tweet_topics)) == 1:
        # if yes, add this index to the list of indices to remove
        indices_to_remove.append(index)
        continue
    # add the tweet_topics list to the tweets_topics list
    tweets_topics.append(tweet_topics)

# remove the rows with the same topic distribution
tweets_topics_df = tweets_df.drop(indices_to_remove)

# create a new dataframe with the topic probabilities for each tweet
topics_df = pd.DataFrame(tweets_topics, columns=[f"Topic {i}" for i in range(lda_model.num_topics)])
# concatenate the tweets dataframe and the topics dataframe
tweets_topics_df.reset_index(drop=True, inplace=True)
topics_df.reset_index(drop=True, inplace=True)
tweets_topics_df = pd.concat([tweets_topics_df, topics_df], axis=1)

# assign the topic with the highest probability to each tweet
tweets_topics_df['Dominant Topic'] = tweets_topics_df.iloc[:, 1:].idxmax(axis=1)

```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

# Define a function to clean the tweets for VADER sentiment analysis
def clean_tweet_for_vader(tweet):
    #tweet = tweet.lower()
    tweet = contractions.fix(tweet)
    tweet = re.sub(r'^b(?:"|\')?(.*)$', r'\1', tweet)
    tweet = re.sub(r"\\n", '', tweet)
    # Remove URLs
    tweet = re.sub(r'http\S+', '', tweet)
    # Remove mentions
    tweet = re.sub(r'@\s*\w+', '', tweet)
    # Remove hashtags
    tweet = re.sub(r'#\s*\w+', '', tweet)
    # Remove numbers
    tweet = re.sub(r'\d+', '', tweet)
    # Remove extra whitespace
    tweet = tweet.strip()
    tweet = ' '.join([word for word in tweet.split() if word not in stop_words or word in ["no", "not","nor","but"]])
    # Remove punctuation from the tweet, except for '!' and '?'
    translator = str.maketrans('', '', string.punctuation.replace('!', '').replace('?', ''))
    tweet = tweet.translate(translator)
    
    return tweet

# Clean the tweets for VADER sentiment analysis
tweets_clustered_vader['clean_tweet'] = tweets_clustered_vader['tweet'].apply(clean_tweet_for_vader)

# AFTER TWEETS ARE CLEANED: Remove duplicate tweets after cleaning and keep only the first occurrence
tweets_clustered_vader = tweets_clustered_vader.drop_duplicates(subset='clean_tweet', keep='first')

# Define a function to apply VADER 
def get_vader_scores(tweet):
    scores = analyzer.polarity_scores(tweet)
    pos_score = scores['pos']
    neu_score = scores['neu']
    neg_score = scores['neg']
    comp_score = scores['compound']
    if comp_score >= 0.05:
        sentiment = 'positive'
    elif comp_score <= -0.05:
        sentiment = 'negative'
    else:
        sentiment = 'neutral'
    return pos_score, neu_score, neg_score, comp_score, sentiment

tweets_clustered_vader['pos'], tweets_clustered_vader['neu'], tweets_clustered_vader['neg'], tweets_clustered_vader['compound'], tweets_clustered_vader['sentiment'] = zip(*tweets_clustered_vader['clean_tweet'].apply(get_vader_scores))

# Output the results in a new dataframe
output_df = tweets_clustered_vader[['username', 'tweet', 'Dominant Topic', 'cluster', 'pos', 'neu', 'neg', 'compound',"sentiment"]]

# Display the output dataframe
output_df.head(10)
```
